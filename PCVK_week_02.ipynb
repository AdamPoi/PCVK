```python
r,g,b = cv.split(img_test)
zg = np.zeros_like(g) # zero green
zr=np.zeros_like(r) # zero red
rb=cv.merge((r,zg,b))
gb=cv.merge((zr,g,b))
final_frame = cv.hconcat((rb, gb))
plt.title('Kiri channel rb, kanan channel gb')
plt.imshow(final_frame)
```
